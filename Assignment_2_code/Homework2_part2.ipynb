```
def zerolistmaker(n):
    listofzeros = [0] * n
    return listofzeros
```

```
a = np.random.uniform(0,1,1000)*100 # from 0 to 100 sample 1000 numbers randomly from a uniform dist. - fruit determination 
b = np.random.uniform(0,1,1000)*100 # from 0 to 100 sample 1000 numbers randomly from uniform dist. - state determination
fruits = zerolistmaker(1000) # list of fruits, 1 = juju, 2 = mongo, 3 = chakava
state = zerolistmaker(1000) # list of fruit states (ripe, unripe), 0 = unripe, 1 = ripe
wl = zerolistmaker(1000) # list of wavelengths (emitted by each fruit)
ripe_probs = zerolistmaker(1000) # list of ripe probabilities (of each fruit being ripe)
monkey_picks = zerolistmaker(1000) # list of fruits that the monkey picks, 1 = picked
```

```
juju_std = 50
mongo_std = 20
chak_std = 100

# def mean for different states
juju_ripe = 600
juju_unripe = 500
mongo_ripe = 580
mongo_unripe = 520
chak_ripe = 400
chak_unripe = 550
```

```
# function to calculate the probability of a fruit reflecting light in a
# given wavelength spectrum being ripe
# applying Bayes
```

```
def probRipe(wl_ceil, wl_floor, fruit):
    #standard deviations
    juju_std = 50
    mongo_std = 20
    chak_std = 100

    juju_ripe = 600
    juju_unripe = 500
    mongo_ripe = 580
    mongo_unripe = 520
    chak_ripe = 400
    chak_unripe = 550

    if fruit == 1: # juju
        mean_ripe = juju_ripe
        mean_unripe = juju_unripe
        std = juju_std
        ripe = 0.1
    elif fruit == 2: # mongo
        mean_ripe = mongo_ripe
        mean_unripe = mongo_unripe
        std = mongo_std
        ripe = 0.5
    else:
        mean_ripe = chak_ripe
        mean_unripe = chak_unripe
        std = chak_std
        ripe = 0.4


    p_ripe = norm.cdf((wl_ceil-mean_ripe)/std) - norm.cdf((wl_floor-mean_ripe)/std) # normcdf is the same as getting the area under the curve to the left of the z-score
    p_unripe = norm.cdf((wl_ceil-mean_unripe)/std) - norm.cdf((wl_floor-mean_unripe)/std) # normcdf is the same as getting the area under the curve to the left of the z-score
    prob = (p_ripe*ripe)/(p_ripe*ripe+p_unripe*(1-ripe))
    return prob
```

```
for idx in range(len(a)):
  if 0 < a[idx] < 10: # fruit is juju, 10% probability range
    fruits[idx] = 1;
    val2 = b[idx] # determine state
    if 0 <= val2 < 15: # ripe juju
        state[idx] = 1
        wl[idx] = generateWl(juju_ripe, juju_std) # generate juju with ripe wl
    else:
        wl[idx] = generateWl(juju_unripe, juju_std) # generate juju with unripe wl
  elif 10 <= a[idx] < 60 : # fruit is mongo
      fruits[idx] = 2
      val2 = b[idx]; # determine state
      if 0 <= val2 < 80: # ripe mongo
          state[idx] = 1
          wl[idx] = generateWl(mongo_ripe, mongo_std) # generate mongo with ripe wl
      else:
          wl[idx] = generateWl(mongo_unripe, mongo_std) # generate mongo with unripe wl
  else: # fruit is chakava, 40# prob range
      fruits[idx] = 3
      val2 = b[idx] # determine state
      if 0 <= val2 < 10: # ripe chakava
          state[idx] = 1
          wl[idx] = generateWl(chak_ripe, chak_std) # generate chakava with ripe wl
      else:
          wl[idx] = generateWl(chak_unripe, chak_std) # generate chakava with unripe wl
```

```
for idx in range(len(wl)):
    val = wl[idx]
    val_ceil = val + 5 # the monkey can identify the wavelength with +/- 5 nm accuracy,
    val_floor = val - 5 # the monkey can identify the wavelength with +/- 5 nm accuracy, 
    ripe_probs[idx] = probRipe(val_ceil, val_floor, fruits[idx]) # get probability of fruit being ripe
    if ripe_probs[idx] >= 0.5: 
        monkey_picks[idx] = 1
```

```
# how many ripe fruits did the monkey pick?
# how many ripe fruits did the monkey leave?
# how many unripe fruits did the monkey pick?
# how many unripe fruits did the monkey leave?
ripes = state.count(1)
unripes = state.count(0)
ripe_picks = 0
ripe_leave = 0
unripe_picks = 0
unripe_leave = 0
for idx in range(len(fruits)):
  if state[idx] == 1 and monkey_picks[idx] == 1:
      ripe_picks = ripe_picks + 1

for idx in range(len(fruits)):
  if state[idx] == 1 and monkey_picks[idx] == 0:
      ripe_leave = ripe_leave + 1

for idx in range(len(fruits)):
  if state[idx] == 0 and monkey_picks[idx] == 1:
      unripe_picks = unripe_picks + 1

for idx in range(len(fruits)):
  if state[idx] == 0 and monkey_picks[idx] == 0:
      unripe_leave = unripe_leave + 1
```
